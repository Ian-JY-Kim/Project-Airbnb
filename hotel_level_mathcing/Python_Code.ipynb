## 1.전처리

<pre>

1. master: hotel_population_20211102_493.dta
 --> 전화 번호 및 호텔 이름 공백 제거
 
2. target: korea_hotel_govnt.csv
 --> 전화 번호 및 호텔 이름 공백 제거
 
</pre>

```python
master = pd.read_stata("hotel_population_20211102_493.dta")
master_simple = master[['id_regist','phone','hotelname', 'sigungu']]

temp_phone = [a.replace(" ", "") for a in master_simple['phone'].tolist()]
master_simple['phone'] = temp_phone

temp_name = [a.replace(" ", "") for a in master_simple['hotelname'].tolist()]
master_simple['hotelname'] = temp_name

master_simple.head(5)
```

```python
target = pd.read_csv("korea_hotel_govnt.csv")
target_simple = target[['결정 등급', '호텔명', '전화번호', '주소']]

temp_phone2 = [a.replace("-", "") for a in target_simple['전화번호'].tolist()]
target_simple['전화번호'] = temp_phone2

temp_name2 = [a.replace(" ", "") for a in target_simple['호텔명'].tolist()]
target_simple['호텔명'] = temp_name2

target_simple = target_simple.rename({'전화번호':'phone'}, axis = 'columns')
target_simple = target_simple.rename({'호텔명':'hotelname'}, axis = 'columns')

target_simple.head(5)
```

<pre>

```python
test_to_concat = test[['id_regist', 'hotelname_x', '결정 등급']]
test_to_concat = test_to_concat.rename({'hotelname_x':'hotelname'}, axis = 'columns')
```

```python
test2_to_concat = test2[['id_regist', 'hotelname', '결정 등급']]
```

```python
merged_by_1and2 = pd.concat([test_to_concat, test2_to_concat], ignore_index=True)
```

```python
merged_by_1and2 = merged_by_1and2.drop_duplicates(subset=['id_regist'])
```

```python
erase = [43136, 6094, 33536, 51416]
to_erase = merged_by_1and2.query('id_regist ==@ erase')
to_erase.index.tolist()
```

```python
merged_by_1and2 = merged_by_1and2.drop(to_erase.index)
merged_by_1and2.reset_index(drop = True, inplace = True)
merged_by_1and2
```

## 6. Method (4) Naver Scraping 

```python
import requests
from bs4 import BeautifulSoup
```

```python
hotel_446 = pd.read_csv("hotel_left446.csv")
```

```python
hotel_list = hotel_446['hotelname'].tolist()
```

```python
url_list = ["https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=" + i for i in hotel_list]
```

```python
url_list
```

```python
rate = []
for url in url_list:
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    div_hotel = soup.find('div', attrs={'class':"nx_hotel sc_new"})
    
    if div_hotel == None:
        rate.append("None")
    
    elif len(div_hotel) == 1:
        div_grade = div_hotel.find('span', attrs={'class':"grade"})
        if div_grade == None:
            rate.append("None")
        else:
            temp = div_grade.get_text()
            rate.append(temp)

    else:
        rate.append("Multiple")
```

```python
len(url_list)
```

```python
len(rate)
```

```python
hotel_446['naver_rate'] = rate
```

```python
hotel_446.to_csv("hotel_446_naver.csv")
```
